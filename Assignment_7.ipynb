Q1. What is the purpose of the try statement?
-->
try - except statements used for to handle exceptions. 
If any line in code raised exception may interrupt the program flow & terminate program, so to avoid this try - except statement.
In case of any statement should execute after successful execution of try statement, try - except - else is used, where in case if any statement should execute after exception occured , try - except - finally is used.



Q2. What are the two most popular try statement variations?
-->
2 try statement variations are:
  1. try - except - else 
     ->In case of any statement should execute after successful execution of try statement, try - except - else is used.
  2. try - except - finally
     -> In case if any statement should execute after exception occured , try - except - finally is used

Q5. What is the purpose of the with/as argument, and what other statement is it like?
--> 
with is used for efficience & clean code in file operations 
To allocate & release memory resources with / as keyword is used
Code without with/as :

f.write('a')
f.close
  This can be replaced using
  with open('filename.txt','w') as f:
