5. What is transferred to the keyword arguments allowRegexes and blockRegexes?
--> Regex patterns are transferred to allowRegexes & blockRegexes.

    Pattern that not to be allowed is transferred to blockRegexes.
    eg. inputStr( blockRegexes = 'aeiou') 
    String containing 'aeiou' wont be allowed
