17. How do you distinguish between copy.copy() and copy.deepcopy()?
--> 
copy.copy() constructs compound object and then  inserts references into it to the objects found in the original.
If changes in made in either original or copied object same changes reflect in other

A deep copy constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original.
If changes made in copied object does not reflect in original one.
